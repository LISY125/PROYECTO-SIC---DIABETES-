```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Data_Final=pd.read_csv("Data_final.csv")
Data_Final.head()
```

```python
x=Data_Final["Diabetes"].value_counts()

y=1,2,3
plt.bar(y,x, color=("lightblue","lightcoral","lightgreen"))
plt.title("Grafica de Poblacion")
plt.xlabel("No Diabetes                    Diabetes                       Prediabetes")
plt.ylabel("Cantidad de Respuestas")
 
plt.show()
```

```python
Data_diabetes=Data_Final[Data_Final["Diabetes"]>0]
Data_diabetes.reset_index(drop=True,inplace=True)
Data_diabetes.head()
```

```python
Edades=Data_diabetes["Edad"].replace({1:"18-24",2:"24-28",3:"28-34",4:"34-38",5:"38-44",
                                   6:"44-48",7:"48-54",8:"54-58",9:"58-64",10:"64-68",11:"68-74",12:"74-78",13:"80 o mas"})

y=Edades.value_counts()
y.sort_index( ascending=True,inplace=True)
y
```

```python
rangos_edad=("18-24","24-28","28-34","34-38","38-44",
            "44-48","48-54","54-58","58-64","64-68","68-74","74-78","80 o mas")

Valores=pd.DataFrame({"Poblacion":y})
Valores
plt.figure(figsize=(10, 6))
plt.bar(rangos_edad, Valores["Poblacion"])
plt.title('Edades mas propensas a la Diabetes')
plt.xlabel('Rango de Edad')
plt.ylabel('Número de Personas')
plt.xticks(rotation=45) 
plt.tight_layout() 
plt.show()
```

```python
Genero=Data_diabetes["Genero"].replace({0:"Femenino",1:"Masculino"})
Ordenado=Genero.value_counts()
Ordenado.sort_index()
plt.pie(Ordenado,labels=Ordenado.index,autopct='%1.1f%%',startangle=80,colors=["lightcoral","skyblue"])
plt.title("Género")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(Data_diabetes["IMC"], bins=100,edgecolor="black")
plt.show()
```

```python
data_0=Data_Final[Data_Final["Diabetes"]==0]
data_1_2=Data_Final[Data_Final["Diabetes"]!=0]

muestra_0=data_0.sample(n=29000, random_state=1)
data_reducida=pd.concat([muestra_0,data_1_2])
data_reducida.head()
```

```python
tabla_contingencia=pd.crosstab(data_reducida["Diabetes"], data_reducida["Hipertension"])

tabla_contingencia.plot(kind="bar", stacked=False)
plt.title('Relación entre Diabetes e Hipertensión')
plt.xlabel("No Diabetes                  Prediabetes                    Diabetes")
plt.ylabel("Número de Casos")
plt.legend(title="Hipertensión", labels=["No Hipertensión", "Hipertensión"])
plt.show()
```

```python
data = pd.read_csv('Data_final.csv')

diabetes_counts = data['Diabetes'].value_counts().to_dict()

hipertension_bins = [0, 120, 140, 160, 180, 200]
data['Hipertension_categoria'] = pd.cut(data['Hipertension'], bins=hipertension_bins)

contar_hipertension = data.groupby(['Hipertension_categoria', 'Diabetes']).size().unstack(fill_value=0)

contar_hipertension.plot(kind='bar', stacked=True)
plt.title('Conteo de personas con hipertensión por estado de diabetes')
plt.xlabel('Categoría de Hipertensión')
plt.ylabel('Número de Personas')
plt.legend(title='Estado de Diabetes', labels=["Sin Diabetes",'Prediabetes', 'Diabetes'])
plt.grid()
plt.show()
```

```python
datos_combinados = data_reducida.groupby(['Colesterol Alto', 'Diabetes']).size().unstack(fill_value=0)

fig, ax = plt.subplots(figsize=(12, 8))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c'] 
datos_combinados.plot(kind='bar', stacked=True, ax=ax, color=colors)

plt.title('Conteo de Personas con Diabetes y Prediabetes dentro de cada Categoría de Colesterol Alto', fontsize=18)
plt.xlabel("No Colesterol Alto                                                            Colesterol Alto",fontsize=14)
plt.ylabel('Número de Personas', fontsize=14)
plt.xticks(rotation=0)

for i, col in enumerate(datos_combinados.index):
    total = datos_combinados.sum(axis=1)[i]
    ax.text(i, total + 50, f'N={total}', ha='center', fontsize=12, color='black')

for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}', (p.get_x() + p.get_width() / 2., p.get_y() + height / 2.),
                    ha='center', va='center', fontsize=12, color='white')

plt.legend(title='Condición', labels=['Sin Diabetes', 'Prediabetes', 'Diabetes'], fontsize=12, loc='upper left')

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
