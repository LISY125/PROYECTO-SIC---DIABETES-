```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Data_completa=pd.read_csv("diabetes_012_health_indicators_BRFSS2015.csv")
Data_completa.columns=["Diabetes","Hipertension","Colesterol Alto",
                       "Control de Colesterol","IMC","Fumador","Derrame Cerebral",
                       "Enfermedad Cardiaca","Actividad Fisica","Comer Frutas","Comer vegetales",
                       "Alto Consumo de Alcohol","Poliza","No Medico","Salud General",
                       "Salud Mental","Salud Fisica","Dificultades para caminar","Genero","Edad","Educacion","Ingreso"]

Data_completa.info()
```

```python
Data_completa.isna().sum()
```

```python
Data_completa.isnull().sum()
```

```python
Duplicados=Data_completa.duplicated()
Duplicados.value_counts()
```

```python
Data_completa.drop_duplicates(inplace=True)
Data_completa.head()
```

```python
Data_real=Data_completa[Data_completa["IMC"]<70]
Data_real.head()
```

```python
Data_final=Data_real[Data_real["IMC"]>15]
Data_final.head()
```

```python
del Data_final["Educacion"]
Data_final.head()
```

```python
Data_final=Data_final.astype(int)
Data_final.head()
```

```python
x=Data_final["Diabetes"].value_counts()
x
```

```python
y=1,2,3
plt.bar(y,x, color=("skyblue","red","orange"))
plt.title("Grafica de Poblacion")
plt.xlabel("No Diabetes                    Diabetes                       Prediabetes")
plt.ylabel("Cantidad de Respuestas")
 
plt.show()
```

```python
Data_diabetes=Data_final[Data_final["Diabetes"]>0]
Data_diabetes.reset_index(drop=True,inplace=True)
Data_diabetes.head()
```

```python
Data_fumadores=Data_diabetes[Data_diabetes["Fumador"]>0]
Data_fumadores.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Assuming 'Data_fumadores' contains your data

# Count the occurrences of each health condition
Hipertension_count = Data_fumadores['Hipertension'].sum()
Colesterol_count = Data_fumadores['Colesterol Alto'].sum()
diabetes_count = len(Data_fumadores)  # All rows in Data_fumadores have Diabetes > 0


# Create a bar chart
conditions = ['Diabetes', 'Hipertension', 'Colesterol Alto']
counts = [diabetes_count, Hipertension_count, Colesterol_count]

plt.bar(conditions, counts, color=['blue', 'orange', 'red'])
plt.xlabel('Health Conditions')
plt.ylabel('Number of Smokers with Diabetes')
plt.title('Health Conditions among Diabetic Smokers')
plt.show()
```
